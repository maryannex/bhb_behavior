```python
#imports
import matplotlib.pyplot as plt
import numpy as np
import array as arr
from matplotlib.ticker import MultipleLocator, FormatStrFormatter, FuncFormatter
```

```python
#simulation from manual
RXTE_HR = np.array([])
Nustar_HR = np.array([])

with open('nustar_HR') as f:
    Nustar_HR = f.readlines()
Nustar_HR = np.array([float(x.strip()) for x in Nustar_HR])

with open('RXTE_HR') as f:
    RXTE_HR = f.readlines()
RXTE_HR = np.array([float(x.strip()) for x in RXTE_HR])
```

```python
#plot
fig1 = plt.figure(figsize=(10,10)) #creates 10x10 size graph
ax1 = fig1.add_subplot(1,1,1) #1:1 grid, 1st subplot 
plt.title('HR of Nustar vs RXTE',fontsize=30)

#x-axis
ax1.set_xscale('linear')
ax1.set_xlabel(r'Nustar', fontsize = 20 )


#y-axis
ax1.set_yscale('linear')
ax1.set_ylabel(r'RXTE', fontsize = 20 )


# #graph
ax1.scatter(Nustar_HR, RXTE_HR, marker='o', color='red')
m1,b1 = np.polyfit(Nustar_HR, RXTE_HR, 1)
ax1.plot(Nustar_HR, m1*Nustar_HR + b1, color = 'red')
```

```python
HR_equation = 'y = ' + str(round(m1,4)) + 'x' ' + ' + str(round(b1,4))
print(HR_equation)
```

```python
with open('nustar_overall_counts') as f:
    Nustar_overallCounts = f.readlines()
Nustar_overallCounts = np.array([float(x.strip()) for x in Nustar_overallCounts])

with open('RXTE_overall_counts') as f:
    RXTE_overallCounts = f.readlines()
RXTE_overallCounts = np.array([float(x.strip()) for x in RXTE_overallCounts])
```

```python
#plot
fig = plt.figure(figsize=(10,10)) #creates 10x10 size graph
ax = fig.add_subplot(1,1,1) #1:1 grid, 1st subplot 
plt.title('Overall Counts of Nustar vs RXTE',fontsize=30)

#x-axis
ax.set_xscale('linear')
ax.set_xlabel(r'Nustar', fontsize = 20 )


#y-axis
ax.set_yscale('linear')
ax.set_ylabel(r'RXTE', fontsize = 20 )


#graph
plt.scatter(Nustar_overallCounts, RXTE_overallCounts, marker='o', color='red')
m,b = np.polyfit(Nustar_overallCounts, RXTE_overallCounts, 1)
ax.plot(Nustar_overallCounts, m*Nustar_overallCounts + b, color = 'red')
```

```python
counts_equation = 'y = ' + str(round(m,4)) + 'x' ' + ' + str(round(b,4))
print(counts_equation)
```
